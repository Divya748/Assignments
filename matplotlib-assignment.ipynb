```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
url="https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
```

```python
titanic = pd.read_csv(url)
```

```python
'''
Create a pie chart presenting the male/female proportion
'''

fig,ax1=plt.subplots(1,1,figsize=(6,6))
titanic.groupby("sex").size().plot(kind="pie",autopct="%.1f%%",colors=["blue","orange"],ax=ax1)
ax1.set_title("male/female proportion")
```

```python
'''
Create a scatterplot with the Fare paid and the Age, differ the plot color by gender
'''
fig,ax2=plt.subplots(1,1,figsize=(6,6))
colors={"male":"green","female":"red"}
ax2.scatter(titanic['fare'].iloc[:-1],titanic['age'].iloc[:-1], c=titanic['sex'].iloc[:-1].map(colors))
ax2.set_xlabel('fare')
ax2.set_ylabel('age')
```
